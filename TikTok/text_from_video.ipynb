```python
import os
import speech_recognition as sr
```

```python
command2mp3 = "ffmpeg -i video_data/tik_tok_meloni.mp4 video_data/tik_tok_meloni.mp3 -y" #first convert to mp3 then to wav using ffmpeg
command2wav = "ffmpeg -i video_data/tik_tok_meloni.mp3 video_data/tik_tok_meloni.wav -y"
```

```python
os.system(command2wav) #ffmpeg command needed (on linux: "sudo apt install ffmpeg")
```

```python
recognizer_instance = sr.Recognizer()


wav = sr.AudioFile("video_data/tik_tok_meloni.wav") # possible formats: .aiff .flac .wav, video downloaded using an online website

with wav as source:
    recognizer_instance.pause_threshold = 3.0
    audio = recognizer_instance.listen(source)
    print("Ok! sto ora elaborando il messaggio!")
try:
    text = recognizer_instance.recognize_google(audio, language="it-IT")
    print("Google ha capito: \n", text)
except Exception as e:
    print(e)
```
